```python
# import important stuff
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
### -----------------------------------------------------------------------------------------------------------###
#                                              Perfect linear Data                                               #
###------------------------------------------------------------------------------------------------------------###
```

```python
# draw random independent variable values
x = np.random.randint(100,size=100)
# set dependent variable values
y = x
print()

# draw random independent variable values
x_test = np.random.randint(100,150,size=50)
# set dependent variable values
y_test = x_test

# to see format of the data uncomment line below
# print(x[:10],'\n',y[:10],'\n',x_test[:10],'\n',y_test[:10])
```

```python
# draw a scatter plot of the x,y and x_test,y_test
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot()

ax.scatter(x,y,label='training data')
ax.scatter(x_test,y_test,label='testing data')
plt.legend()
```

```python
# create a suitable format for fit() function
# (just combining points x,y and x_test,y_test)
points = []
for i in range(len(x)):
    points.append([x[i],y[i]])
points = np.array(points)

test = []
for i in range(len(x_test)):
    test.append([x_test[i],y_test[i]])
test = np.array(test)

# if you want to see the point coordinate format uncomment line below
# print(points[:10],'\n\n',test[:10])
```

```python
reg = linear_model.LinearRegression()
reg.fit(points,y)
preds = reg.predict(test)
preds = np.array(preds)

# to see how predictions work uncomment line below
# print(preds.astype(int))
```

```python
# draw a scatter plot of the x,y and x_test,y_test
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot()

ax.plot(preds,y_test,label='fit',c='green')
ax.scatter(points[:,0],points[:,1],label='training data')
ax.scatter(test[:,0],test[:,1],label='test data')

plt.legend()
```
